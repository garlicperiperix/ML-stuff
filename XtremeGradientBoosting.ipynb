```python
import numpy as np

import numpy
import pandas as pd

from numpy import loadtxt
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from xgboost import plot_tree
#import matplotlib.pyplot as plt
# load data
dataset=pd.read_csv("D:/Data.csv")
```

```python
del dataset['commitid']
del dataset['revid']
del dataset['totaltimedeficit']
del dataset['alerts']
del dataset['inconsistency']
del dataset['curfew']

del dataset['paxcapacity']
del dataset['duplicate']
```

```python
dataset
```

```python
# split data into X and y

X= dataset.iloc[:,0:8]
Y= dataset.iloc[:,8:9]
```

```python
dataset.shape
```

```python
# split data into train and test sets


X_train, X_test, Y_train, Y_test = train_test_split(X, Y, random_state=0, train_size= 0.9 )

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)



# fit model no training data
model = XGBClassifier(max_depth="20", n_estimators = 1000, learning_rate= 0.5)
model.fit(X_train, Y_train)
# make predictions for test data
y_pred = model.predict(X_test)
predictions = [round(value) for value in y_pred]
# evaluate predictions
accuracy = accuracy_score(Y_test, predictions)
print("Accuracy: %.2f%%" % (accuracy * 100.0))
print('The accuracy of the xgb classifier on test data is {:.2f} out of 1 on testing data'.format(model.score(X_test, Y_test)))
print('The accuracy of the xgb classifier is {:.2f} out of 1 on training data'.format(model.score(X_train, Y_train)))
```
